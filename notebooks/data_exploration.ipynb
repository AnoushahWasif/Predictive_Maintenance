```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load raw data
data = pd.read_csv('../data/raw/equipment_data.csv')
```

```python

# Explore data
print(data.head())
print(data.describe())
```

```python

# Plot feature distributions
data['temperature'].plot(kind='hist', title='Temperature Distribution')
plt.xlabel('Temperature')
plt.show()
```

```python

data['pressure'].plot(kind='hist', title='Pressure Distribution')
plt.xlabel('Pressure')
plt.show()
```

```python

data['vibration'].plot(kind='hist', title='Vibration Distribution')
plt.xlabel('Vibration')
plt.show()
```
