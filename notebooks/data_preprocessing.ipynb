```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


```

```python
# Load raw data
data = pd.read_csv('../data/raw/equipment_data.csv')

```

```python
# Preprocess data
data.fillna(method='ffill', inplace=True)
scaler = StandardScaler()
data[['temperature', 'pressure', 'vibration']] = scaler.fit_transform(data[['temperature', 'pressure', 'vibration']])

```

```python
# Save processed data
data.to_csv('../data/processed/equipment_data_processed.csv', index=False)
```
